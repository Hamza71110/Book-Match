## **Popularity-based-Recommender**

```python
ratings_with_name = ratings.merge(books, on='ISBN')
```

```python
num_rating_df=ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_ratings'}, inplace=True)
num_rating_df
```

```python
avg_rating_df=ratings_with_name.groupby('Book-Title')['Book-Rating'].mean().reset_index()
avg_rating_df.rename(columns={'Book-Rating':'avg_rating'}, inplace=True)
avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df, on='Book-Title')
popular_df
```

```python
popular_df = popular_df[popular_df['num_ratings'] >=100].sort_values('avg_rating', ascending=False).head(50)
popular_df.shape
```

```python
popular_df=popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Year-Of-Publication','Image-URL-M','num_ratings','avg_rating']]
```

## **Collaborative-Filtering-Based-Recommender-System**

```python
x = ratings_with_name.groupby('User-ID').count()['Book-Rating']>100
selected_users = x[x].index
```

```python
selected_users.shape
```

```python
filtered_rating=ratings_with_name[ratings_with_name['User-ID'].isin(selected_users)]
```

```python
filtered_rating.shape
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>30
famous_books = y[y].index
```

```python
famous_books.shape
```

```python
final_rating=filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_rating
```

```python
pt=final_rating.pivot_table(index='Book-Title', columns='User-ID', values='Book-Rating')
```

```python
pt
```

```python
pt.shape
```

```python
pt.isnull().sum()
```

```python
pt.fillna(0, inplace=True)
```

```python
pt.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarities = cosine_similarity(pt)
```

```python
similarities.shape
```

```python
similarities
```

```python
# Function for recommendations
def recommend(book_name):
    # fecthing index of user selected movie
    index = np.where(pt.index == book_name)[0][0]
    similar_movies = sorted(list(enumerate(similarities[index])),key=lambda x:x[1],reverse=True)[1:9]

    data = []
    for i in similar_movies:
        item = []
        temp_df = books[books['Book-Title']== pt.index[i[0]]]
        item.extend(list(temp_df['Book-Title'].values))
        item.extend(list(temp_df['Book-Author'].values))
        item.extend(list(temp_df['Image-URL-M'].values))
        item.extend(list(temp_df['Year-Of-Publication'].values))



        data.append(item)

    return data

```

```python
recommend('2nd Chance')
```

```python
import pickle
pickle.dump(popular_df, open('popular_df.pkl', 'wb'))
pickle.dump(pt, open('pt.pkl', 'wb'))
pickle.dump(books, open('books.pkl', 'wb'))
pickle.dump(similarities, open('similarities.pkl', 'wb'))
pickle.dump(recommend, open('recommend.pkl', 'wb'))
```
